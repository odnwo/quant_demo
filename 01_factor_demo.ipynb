```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 构造模拟数据（你之后会换成真实数据）
np.random.seed(42)
dates = pd.date_range("20220101", periods=100)
price = np.cumsum(np.random.randn(100)) + 100
df = pd.DataFrame({"Close": price}, index=dates)

# 添加简单因子：5日、20日均线
df["MA5"] = df["Close"].rolling(5).mean()
df["MA20"] = df["Close"].rolling(20).mean()

# 可视化
plt.figure(figsize=(12, 5))
df[["Close", "MA5", "MA20"]].plot(ax=plt.gca())
plt.title("Price with MA5 / MA20")
plt.grid(True)
plt.show()
```
